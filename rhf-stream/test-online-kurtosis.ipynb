```python
import scipy.io as sio
import scipy.stats as sstats
import online_kurtosis as oks
import numpy as np
```

```python
mat_contents = sio.loadmat("../datasets/cardio.mat")
dataset = mat_contents['X'] 
```

```python
for i in range(0, dataset.shape[1]):
    print(dataset[:, i])
    print("scipy=", sstats.kurtosis(dataset[:, i], fisher=False))
    print("online=", oks.online_kurtosis(dataset[:, i]))
```
